```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

- ### lets read the data set:

```python
df = pd.read_csv('./googleplaystore.csv')

df.head(5)
```

- ###  Visulize our data as Rating to the Category :

```python
df = df[df['Rating'] <= 5.0]
sns.relplot(y='Rating', x='Category', data=df, height=10);
```
